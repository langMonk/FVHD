## Examples of using IVHD

To execute this notebook IVHD must be installed in your python environment.

#### Local installation
To install `ivhd` package in editable mode run `pip install -e .` in base directory of this repository.

#### PyPI installation
To install `ivhd` package from PyPI run `pip install ivhd`


```python
from ivhd import IVHD, IVHDGrad
from sklearn.datasets import load_digits
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
from matplotlib import pyplot as plt

dataset = load_digits()
dataset
X = dataset["data"]
y = dataset["target"]
X = X / 255.
```

```python
pca = PCA()
X_pca = pca.fit_transform(X)
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=y, s=2)
```

```python
tsne = TSNE()
X_tsne = tsne.fit_transform(X)
plt.scatter(X_tsne[:, 0], X_tsne[:, 1], c=y, s=2)
```

```python
ivhd = IVHD(simulation_steps=1000, nn=5, rn=2, lambda_=0.95, c=0.05)
X_ivhd = ivhd.fit_transform(X)
plt.scatter(X_ivhd[:, 0], X_ivhd[:, 1], c=y, s=2)
```

```python
ivhd_grad = IVHDGrad(steps=2000, nn=5, rn=2, re_draw_remote_neighbors=True, optimizer_params={"lr": 0.005})
X_ivhd_grad = ivhd_grad.fit_transform(X)
plt.scatter(X_ivhd_grad[:, 0], X_ivhd_grad[:, 1], c=y, s=2)
```

```python
plt.show()
```
