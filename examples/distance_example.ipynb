## Examples of using IVHD comparing results in changing distances

To execute this notebook IVHD must be installed in your python environment.

#### Local installation
To install `ivhd` package in editable mode run `pip install -e .` in base directory of this repository.

#### PyPI installation
To install `ivhd` package from PyPI run `pip install ivhd`

```python
from sklearn.datasets import load_digits

dataset = load_digits()
dataset
X = dataset["data"]
y = dataset["target"]
X = X / 255.
```

```python
from ivhd import IVHD
from matplotlib import pyplot as plt

ivhd_euclidean = IVHD(simulation_steps=1000, nn=5, rn=2, lambda_=0.95, c=0.05, distance="euclidean")
X_ivhd_e = ivhd_euclidean.fit_transform(X)
plt.scatter(X_ivhd_e[:, 0], X_ivhd_e[:, 1], c=y, s=2)
```

```python
ivhd_binary= IVHD(simulation_steps=1000, nn=5, rn=2, lambda_=0.95, c=0.05, distance="binary")
X_ivhd_b = ivhd_binary.fit_transform(X)
plt.scatter(X_ivhd_b[:, 0], X_ivhd_b[:, 1], c=y, s=2)
```

```python
ivhd_cosine= IVHD(simulation_steps=1000, nn=5, rn=2, lambda_=0.95, c=0.05, distance="cosine")
X_ivhd_c = ivhd_cosine.fit_transform(X)
plt.scatter(X_ivhd_c[:, 0], X_ivhd_c[:, 1], c=y, s=2)
```

```python
from ivhd import IVHDGrad
ivhd_grad_cosine = IVHDGrad(steps=2000, nn=5, rn=2, re_draw_remote_neighbors=True, optimizer_params={"lr": 0.005}, distance='cosine')
X_ivhd_grad_cosine = ivhd_grad_cosine.fit_transform(X)
plt.scatter(X_ivhd_grad_cosine[:, 0], X_ivhd_grad_cosine[:, 1], c=y, s=2)
```

```python
ivhd_grad_euclidean = IVHDGrad(steps=2000, nn=5, rn=2, re_draw_remote_neighbors=True, optimizer_params={"lr": 0.005}, distance='euclidean')
X_ivhd_grad_euclidean = ivhd_grad_euclidean.fit_transform(X)
plt.scatter(X_ivhd_grad_euclidean[:, 0], X_ivhd_grad_euclidean[:, 1], c=y, s=2)
```

```python
ivhd_grad_binary = IVHDGrad(steps=2000, nn=5, rn=2, re_draw_remote_neighbors=True, optimizer_params={"lr": 0.005}, distance='binary')
X_ivhd_grad_binary = ivhd_grad_binary.fit_transform(X)
plt.scatter(X_ivhd_grad_binary[:, 0], X_ivhd_grad_binary[:, 1], c=y, s=2)
```
